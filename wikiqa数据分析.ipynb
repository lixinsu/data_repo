```python
import os
import sys
import numpy as np
import pandas as pd
import json
import random
```

```python
train_file = '/data/sulixin/research/contest/wikiqa/WikiQACorpus/WikiQA-train.tsv'
dev_file = '/data/sulixin/research/contest/wikiqa/WikiQACorpus/WikiQA-dev.tsv'
test_file = '/data/sulixin/research/contest/wikiqa/WikiQACorpus/WikiQA-test.tsv'
dev_filter = '/data/sulixin/research/contest/wikiqa/WikiQACorpus/'
```

```python
train = pd.read_csv(train_file, sep='\t')
dev = pd.read_csv(dev_file, sep='\t')
test = pd.read_csv(test_file, sep='\t')
```

```python
def func(x):
    rv = {'candidates': x['Sentence'].tolist(), 'query': x['Question'].iloc[0], 
          'labels': x['Label'].tolist(), 'title': x['DocumentTitle'].iloc[0]}
    return pd.Series(rv)

def gt_cnt(li):
    return len([x for x in li if x==1])

def word_len(t):
    return len(t.split())

def avg_word_len(x):
    def avg(li):
        return sum(li) * 1 / len(li)
    return avg([word_len(y) for y in x])
        
        
def my_stat(data):
    df = data.groupby(['QuestionID']).apply(func)
    df['q_len'] = df['query'].apply(word_len)
    df['p_cnt'] = df['labels'].apply(len)
    df['gt_cnt'] = df['labels'].apply(gt_cnt)
    df['p_len'] = df['candidates'].apply(avg_word_len) 
    print(df['gt_cnt'].value_counts().to_dict())
    print(json.dumps(df[df['gt_cnt'] == 2][['query','labels','candidates']].iloc[2].to_dict(), indent=2, ensure_ascii=False))
    return df.describe()

my_stat(train)
```

```python
my_stat(dev)
```

```python
my_stat(test)
```
