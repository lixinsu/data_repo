```python
import os
import sys
import json
import numpy as np
import pandas as pd
```

```python
train_s_file = "/data/sulixin/research/contest/DuReader/data/raw/trainset/search.train.json"
dev_s_file = "/data/sulixin/research/contest/DuReader/data/raw/devset/search.dev.json"
test_s_file = "/data/sulixin/research/contest/DuReader/data/raw/testset/search.test.json"
```

```python
#数据读入
def load_dataset(filename):
    data = []
    for line in open(filename):
        data.append(json.loads(line))
    print( '%s has %s samples' % (os.path.basename(filename),len(data)  ))
    return data
```

```python
train_s = load_dataset(train_s_file)
dev_s = load_dataset(dev_s_file)
test_s= load_dataset(test_s_file)
```

```python

# 数据统计代码

def avg_para(docs):
    paras = [len(x['paragraphs']) for x in docs]
    return sum(paras) * 1.0 / len(paras)

def avg_len(li):
    if len(li) == 0:
        return 0
    data = [len(d) for d in li]
    return sum(data) * 1.0 / len(data)    
        
def sum_para(docs):
    paras = [len(d['paragraphs']) for d in docs]
    return sum(paras)

def len_para(docs):
    paras = [len(p) for d in docs for p in d['paragraphs']]
    return sum(paras) * 1.0 / len(paras)

def count_select(docs):
    return len( [d for d in docs if d['is_selected']] )
        
def avg_title(docs):
    data = [len(d['title']) for d in docs]
    return sum(data) * 1.0 / len(data)
    
def x_hit(docs, idx):
    return int(docs[idx]['is_selected']) if idx < len(docs) else None 
    
def my_stat(data):
    pre_data = {'avg_doc':[len(x['documents']) for x in data],\
                'avg_para_d': [avg_para(x['documents']) for x in data], \
                'sum_para': [sum_para(x['documents']) for x in data], \
                'len_para': [len_para(x['documents']) for x in data],\
                'select': [count_select(x['documents']) for x in data],\
                'len_q':  [len(x['question']) for x in data],\
                'n_answers': [len(x['answers']) for x in data],\
                'len_answers': [avg_len(x['answers']) for x in data],\
                'len_title': [avg_title(x['documents']) for x in data],\
                'q_type': [x['question_type'] for x in data],\
                '1_hit':[x_hit(x['documents'], 0) for x in data],\
                '2_hit':[x_hit(x['documents'], 1) for x in data], \
                '3_hit': [x_hit(x['documents'], 2) for x in data],\
                '4_hit': [x_hit(x['documents'], 3) for x in data]
               }
    train_df = pd.DataFrame(pre_data)
    print('-----overall----')
    print(train_df.describe())
    for qtype in ['ENTITY', 'DESCRIPTION', 'YES_NO']:
        print('-----%s------' % qtype)
        print('percentage: %.4f' % (sum( train_df['q_type']==qtype)/len(train_df)))
        print(train_df[ train_df['q_type']==qtype ].describe())
    print('no answers %s, rate %.4f' % (sum(train_df['select']==0) , sum(train_df['select']==0) * 1.0 / len(train_df)))
```

```python
my_stat(train_s)
```
